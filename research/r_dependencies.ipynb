## R Dependencies

This notebook installs R and the R packages needed to perform all the calculations throughout this project into a project-local sub-folder.

**Note:** This notebook requires sudo privileges to install R and assumes a Ubuntu/Debian based system!

**Important:** Once any notebook first imports anything from the `rpy2` site package, a new R process is spawned off and used to run all calls to R via `rpy2`. In order for this process to use the project-local sub-folder for the R dependencies, an environment variable `R_LIBS` must be set (with the path to this sub-folder) **before** any interaction with `rpy2`.

```python
import getpass
import os
import shutil
```

Obtain the sudo password.

```python
password = getpass.getpass()
```

Install the packages *r-base r-base-dev libcurl4-openssl-dev libxml2-dev patchelf*

```python
os.system(
    f"echo {password} | sudo -S apt-get -y install"
    " r-base r-base-dev libcurl4-openssl-dev libxml2-dev patchelf"
);
```

Create a (hidden) folder in the project's root directory to install the R libraries into.

```python
%cd ..
```

```python
# Sanity check to see if the R_LIBS variable is set.
assert os.getenv("R_LIBS")
# Expand the R_LIBS path to an absolute path.
r_libs_path = os.path.join(os.getcwd(), os.environ["R_LIBS"])
```

```python
# Create a fresh folder for the R dependencies.
try:
    shutil.rmtree(r_libs_path)
except FileNotFoundError:
    pass
os.mkdir(r_libs_path)
```
